**2-Conversor de Temperatura**

```python
def celsius_para_fahrenheit(celsius):
    fahrenheit = celsius * (9/5) + 32
    return fahrenheit

temp_c = 25
temp_f = celsius_para_fahrenheit(temp_c)
print(temp_c, "°C equivalem a", temp_f, "°F")
```

**4-Quadrado e Cubo**

```python
def calcular_potencias(n):
    quadrado = n ** 2
    cubo = n ** 3
    return quadrado, cubo

result = calcular_potencias(4)
print("O quadrado e o cubo de 4 são:", result)
```

**6-Calculo de Desconto**

```python
def calcular_desconto(valor_original, percentual_desconto):
    valor_final = valor_original * (1 - (percentual_desconto / 100))
    return valor_final

preco = 100
desconto = 20
result = calcular_desconto(preco, desconto)
print("O valor final com desconto é:", result)
```

**7-Maior e Menor**

```python
def encontrar_extremos(lista):
    maior = max(lista)
    menor = min(lista)
    return maior, menor

numeros = [4, 9, 1, 7, 3]
result = encontrar_extremos(numeros)
print("Maior e menor número da lista são:", result)
```

**8-Calculadora Simples**

```python
def calculadora(a, b, operacao):
    if operacao == "soma":
        return a + b
    elif operacao == "subtracao":
        return a - b
    elif operacao == "multiplicacao":
        return a * b
    elif operacao == "divisao":
        if b != 0:
            return a / b
        else:
            return "Erro: divisão por zero não é permitida."
    else:
        return "Operação inválida."

result = calculadora(10, 2, "divisao")
print("Resultado:", result)
```

**9-Mensagem Automática**

```python
def enviar_email(destinatario, assunto, corpo="Este é um email automático."):
    print("Enviando email para:", destinatario)
    print("Assunto:", assunto)
    print("Corpo:", corpo)

enviar_email("exemplo@email.com", "Aviso Importante")
enviar_email("usuario@email.com", "Boas-vindas", "Olá! Seja bem-vindo à nossa plataforma.")
```

**10-Sistema de Notas**

```python
def avaliar_aluno(nota):
    if nota >= 7:
        return "Aprovado"
    elif nota >= 5 and nota < 7:
        return "Recuperação"
    else:
        return "Reprovado"

result = avaliar_aluno(6)
print("Situação do aluno:", result)
```
