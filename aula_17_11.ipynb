#Função com retorno e com parâmetro
validar_senha(senha)

Recebe uma string e retorna um dicionário contendo todos os critérios avaliados:

tamanho_valido – senha com 8 ou mais caracteres

tem_numero – contém pelo menos um número

tem_maiuscula – contém pelo menos uma letra maiúscula

tem_minuscula – contém pelo menos uma letra minúscula

tem_especial – contém caractere especial

senha_valida – só é verdadeiro se todos os critérios forem atendidos

```python
def validar_senha(senha):
    criterios = {
        "tamanho_valido": len(senha) >= 8,
        "tem_numero": any(c.isdigit() for c in senha),
        "tem_maiuscula": any(c.isupper() for c in senha),
        "tem_minuscula": any(c.islower() for c in senha),
        "tem_especial": any(c in "!@#$%^&*()-_=+[]{};:,.?/|\\<>~" for c in senha)
    }

    criterios["senha_valida"] = all(criterios.values())
    return criterios


senha_usuario = input("Digite sua senha para validação: ")
resultado = validar_senha(senha_usuario)

print("\nResultado da validação:")
for chave, valor in resultado.items():
    print(f"{chave}: {valor}")

if resultado["senha_valida"]:
    print("\nSenha válida.")
else:
    print("\nSenha inválida.")
```
